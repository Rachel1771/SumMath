```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sup37 = pd.read_csv('supply37.csv')
sup37
```

```python
sup = pd.read_csv('supplyer.csv')
sup['n_id'] = range(1,403)
sup.set_index('n_id', inplace=True)
sup
```

```python
sup37['mat'] = sup.loc[sup37.id]['材料分类'].values
```

```python
sup37.set_index('mat', inplace=True)
sup37
```

```python
sup37.loc['A'] = sup37.loc['A']/0.6
sup37.loc['B'] = sup37.loc['B']/0.66
sup37.loc['C'] = sup37.loc['C']/0.72
```

```python
sup37_array = np.array(sup37.sum())[1:]
sup37_array
```

```python
sup37_array.shape
```

```python
avg_need_eqC = 28200
```

```python
plt.style.use('ggplot')
plt.figure(dpi=200)
need_array_eqC = np.array([avg_need_eqC for i in range(24)])

plt.figure(figsize=[30,4])
plt.xticks(range(1,25),fontsize=10)

plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.yticks(fontsize=10)   
plt.title('选中的37家供应商历史供货量与生产需求关系', fontsize = 20)

plt.bar(x=range(1,25), height=sup37_array,color='darkcyan', label='供货量')
plt.plot(range(1,25), need_array_eqC, label='生产需求', marker='o',linewidth =5.0,markersize=10)
plt.legend(loc='center left',fontsize = 15)

plt.savefig('选中的37家供应商历史周期供货量与生产需求关系.jpg')
```
